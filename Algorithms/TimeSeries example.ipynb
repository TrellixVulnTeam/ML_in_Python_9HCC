```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
%config INlineBackend.figure_format ='retina'
```

```python
btc = web.get_data_yahoo('BTC-USD',start = datetime.datetime(2017,1,1), end = datetime.datetime(2017,12,31))
```

```python
btc.head()
```

```python
btc.tail()
```

```python
plt.figure(figsize=(14,7))
plt.plot(btc['Adj Close'])
```

### Moving Average

```python
rollavg_d3 = btc['Adj Close'].rolling(window=3).mean()
rollavg_d5 = btc['Adj Close'].rolling(window=5).mean()
rollavg_d15 = btc['Adj Close'].rolling(window=15).mean()
```

```python
plt.figure(figsize = (14,7))
plt.plot(btc.index,rollavg_d15,lw=3,alpha=0.8,label = 'Rolling mean 15')
plt.plot(btc.index,rollavg_d3,lw=3,alpha=0.8,label = 'Rolling mean 3')
```
