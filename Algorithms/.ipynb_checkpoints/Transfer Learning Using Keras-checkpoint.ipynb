### Transfer learning

Transfer learning is useful for few reasons:
1. Knowledge gained while solving one problem can be "transfered" to a different problem
2. The traditional cutting edge algorithms needs lots of resources to replicate and build from scratch.
3. Even built from the scratch it is very hard to match the performance of a existing algorithms (eg: VGG, ResNet)


```python
import numpy as np
import os
from keras.models import Sequential, Model
from keras.layers import Input,Dropout, Flatten, Dense,Activation
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.utils import plot_model

from keras.utils import np_utils
from sklearn.utils import shuffle
from sklearn.cross_validation import train_test_split

import matplotlib.pyplot as plt
%matplotlib inline

from keras.applications import vgg19, inception_v3, resnet50
```

### VGG19 Architecture

![alt text](./images/imagenet_vgg16.png "VGG19 Architecture")

### Image preprocessing

```python
img_width, img_height = 224,224
batch_size= 16
epochs = 5

PATH = os.getcwd()
datapath = PATH + '/data/imgdata'
list_img_dir = os.listdir(datapath)
list_img_dir.remove('.DS_Store')
print(list_img_dir)
#plt.imshow(np.uint8(img))
img_data_list = []

for dataset in list_img_dir:
    list_img_names = os.listdir(datapath + '/' + dataset)
    for img in list_img_names:
        img_path = datapath + '/' + dataset + '/' + img
        img = load_img(img_path,target_size=(img_width, img_height))
        img = img_to_array(img)
        #img = np.expand_dims(img,axis=0)
        img_data_list.append(img)

img_data = np.array(img_data_list)
#img_data = np.rollaxis(img_data,1,0)

nclasses = 4
n_samples = img_data.shape[0]
labels = np.ones((n_samples,),dtype = 'int64')
labels[0:202]=0
labels[203:404]=1
labels[405:606]=2
labels[607:]=3

names = list_img_dir
Y = np_utils.to_categorical(labels,nclasses)
```

```python
X,Y = shuffle(img_data,Y,random_state=2)
Xtrain,Xtest, Ytrain,Ytest = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
input_img = Input(shape=(224,224,3))
model = vgg19.VGG19(weights = "imagenet", include_top=False, input_tensor=input_img)
model.summary()
```

### Building custom layers 

```python
last_layer = model.get_layer('block5_pool').output
x = Flatten(name='flatten')(last_layer)
x = Dense(128,activation = 'relu', name = 'fc1')(x)
x = Dense(64, activation = 'relu', name = 'fc2')(x)
out = Dense(nclasses,activation = 'softmax', name = 'output_layer')(x)
custom_vgg_model = Model(input_img,out)

plot_model(custom_vgg_model, to_file="./images/custom_vgg_model.png")
```

![alt text](./images/custom_vgg_model.png "VGG19 Architecture")

```python
for layer in custom_vgg_model.layers[:-3]:
    layer.trainable=False
```

```python
custom_vgg_model.compile(loss='categorical_crossentropy',optimizer ='adam',metrics =['accuracy'])
```

```python
history = custom_vgg_model.fit(Xtrain,Ytrain, batch_size=batch_size,epochs=epochs,verbose=2, validation_data=(Xtest,Ytest))
```

```python
(loss,accuracy) = custom_vgg_model.evaluate(Xtest,Ytest,batch_size=10, verbose=1)
```

```python
print("[INFO] loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))
```

```python
import matplotlib.pyplot as plt
# visualizing losses and accuracy
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_acc = history.history['acc']
val_acc = history.history['val_acc']
xc = range(5)

plt.figure(1,figsize=(7,5))
plt.plot(xc,train_loss)
plt.plot(xc,val_loss)
plt.xlabel('num of Epochs')
plt.ylabel('loss')
plt.title('train_loss vs val_loss')
plt.grid(True)
plt.legend(['train','val'])
#print plt.style.available # use bmh, classic,ggplot for big pictures
plt.style.use(['classic'])

plt.figure(2,figsize=(7,5))
plt.plot(xc,train_acc)
plt.plot(xc,val_acc)
plt.xlabel('num of Epochs')
plt.ylabel('accuracy')
plt.title('train_acc vs val_acc')
plt.grid(True)
plt.legend(['train','val'],loc=4)
#print plt.style.available # use bmh, classic,ggplot for big pictures
plt.style.use(['classic'])
```
