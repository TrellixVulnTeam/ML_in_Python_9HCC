# K Fold cross validation and Grid Search

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, KFold,cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data
y = iris.target
#Splitting data without K Fold cross validation
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,random_state=4, test_size=0.25) 
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(Xtrain,ytrain)
ypred = knn.predict(Xtest)
metrics.accuracy_score(ytest,ypred)
```

### K fold cross validation 

```python
# Example to show training and test splits
knn_cv = KNeighborsClassifier(n_neighbors=5)
scores = cross_val_score(knn_cv,X,y,cv=10)
print(scores, scores.mean())
```

```python
near_n = np.arange(1,31)
accuscore = []
for i in near_n:
    knn_cv = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(knn_cv,X,y,cv=50,scoring = 'accuracy')
    accuscore.append(score.mean())
```

```python
plt.plot(near_n, accuscore)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
krange = np.arange(1,31)
paramgrid = dict(n_neighbors = krange)
knn_gs = KNeighborsClassifier(n_neighbors=5)
grid = GridSearchCV(knn_gs,paramgrid, cv=10,scoring='accuracy',return_train_score=True)
```

```python
grid.fit(X,y)
```

```python
grid.cv_results_['mean_test_score']
```

```python
scores = grid.cv_results_['mean_test_score']
plt.plot(krange,scores)
plt.show()
```

### Iterating through multiple parameters

```python
krange = np.arange(1,31)
weightoption = ['uniform', 'distance']
paramgrid = dict(n_neighbors = krange, weights=weightoption )
gridm = GridSearchCV(knn,param_grid = paramgrid, cv=10,scoring ='accuracy')
gridm.fit(X,y)
```

```python
gridm.best_params_
```

### Reducing computational  using RandomizedSearchCV

```python
from sklearn.grid_search import RandomizedSearchCV
param_dist = dict(n_neighbors=krange,weights=weightoption)
rand = RandomizedSearchCV(knn,param_dist,cv=10,scoring='accuracy')
rand.fit(X,y)
```

```python
rand.best_params_
```
