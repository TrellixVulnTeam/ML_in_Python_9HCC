# Neural Networks in Keras

## Overview
1. Define network -
    Build the skeleton of the network
2. Compile network - 
    It transforms the simple sequence of layers that we defined into a highly efficient           series of matrix transforms in a format intended to be executed on your GPU or CPU,           depending on how Keras is configured.
3. Fit network -
    Adapts the weights on training set
4. Evaluate network
5. Make predictions

```python
from keras.models import Sequential
from keras.layers import  Dense
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./ML in Python/data/pima-indians-diabetes.csv",header=None)
```

```python
Xtrain = df.iloc[:,0:8]
X = np.array(Xtrain)
ytrain = df.iloc[:,8]
y = np.array(ytrain)
```

```python
#Define network
model = Sequential()
model.add(Dense(12,input_dim =8, activation='relu'))
model.add(Dense(1,activation='sigmoid'))

#Compile network
model.compile(loss='binary_crossentropy', optimizer ='adam', metrics=['accuracy'])

# Fit network
history = model.fit(X, y, epochs=100, batch_size=10, verbose=1)

#Evaluate network
loss, accuracy = model.evaluate(X, y)
print("\nLoss: %.2f, Accuracy: %.2f%%" % (loss, accuracy*100))

#Make predictions
probabilities = model.predict(X)
predictions = [float(round(x)) for x in probabilities]
accuracy = np.mean(predictions == y)
print("Prediction Accuracy: %.2f%%" % (accuracy*100))
```
